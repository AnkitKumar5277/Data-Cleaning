```python
df = pd.read_csv("weather_data.csv", parse_dates = ["day"])
df.set_index('day', inplace = True)
df
# day column ko index bana diya gaya aur ab df me day separate column ki jagah index ke roop me show hoga aur bold kardega
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
# missing values ko set kiye gaye default values se replace karke ek naya DataFrame new_df banaya gaya hai.)
```

```python
new_df = df.fillna(method = "ffill")
new_df
 # jo bhi value missing thi, usko uske upar wali value se bhar diya gaya hai.
```

```python
new_df = df.fillna(method = "bfill")
new_df
# jo bhi value missing thi, usko uske niche wali value se bhar diya gaya hai
```

```python
new_df = df.fillna(method="bfill", axis = "columns")
new_df
# df me jo missing values (NaN) hain, unko backward fill karke same row ke agle column ki value se fill kiya gaya hai, aur result new_df me store hua hai.
```

```python
new_df = df.fillna(method="bfill")
new_df
# missing values ko uske niche wale row ki value se bhar diya gaya hai
```

```python
new_df = df.fillna(method = "ffill", limit=1)
new_df
 # har row me ek hi missing value ko uske upar wali value se bhar diya gaya hai.)
```

```python
new_df = df.fillna(method = "ffill", limit=2)
new_df
 # har row me do missing values ko unke upar wali value se bhar diya gaya hai.)
```

```python
new_df = df.interpolate()
new_df
# missing values ko unke aas-paas ke data ke beech ka calculation karke bhar diya gaya hai.)
```

```python
new_df = df.dropna()
new_df
# df me jitni bhi rows me koi bhi missing value (NaN) thi, un sari rows ko hata diya gaya hai — aur jo clean data bacha wo new_df me store hua hai.
```

```python
new_df = df.dropna(how = "all")
new_df
# df me sirf wahi rows delete ki gayi hain jismein saare columns ki values missing (NaN) thi — aur jo baaki clean data bacha, wo new_df me store hua ha
```

```python
new_df = df.dropna(thresh=1)
new_df
 # jisme 1 bhi valid value hai, wo rows bachayi gayi hain, aur jo puri tarah se missing thi wo hata di gayi hain
```

```python
new_df = df.dropna(thresh=2)
new_df
# df me sirf wahi rows rakhi gayi hain jisme kam se kam 2 non-missing (valid) values ho, aur jinki 2 se kam valid values thi — wo rows hata di gayi hain. Result new_df me store hua hai.
```

```python
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
# naye date index ke according df ko set kiya gaya, aur jo dates pehle nahi thi unpe NaN aa jayega
```

```python
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
# df ka index ab 01 Jan 2017 se 11 Jan 2017 tak ka date bana diya gaya, aur jahan pehle data nahi tha, wahan NaN fill ho gaya.
```
